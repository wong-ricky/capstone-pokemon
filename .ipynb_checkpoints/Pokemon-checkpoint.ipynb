### Data Preparation
- Remove mega evolutions

```python
pokemon_clean_df = pokemon_df.copy(deep=True)
pokemon_clean_df
```

```python
# remove mega evolutions
pokemon_clean_df.drop(pokemon_clean_df[pokemon_clean_df['Name'].str.contains('Mega')].index, inplace=True)
pokemon_clean_df
```

```python
# remove pokedex number and name. 
# remove type 2 to simplify 
# remove legendary
pokemon_clean_df.drop(['#','Name','Type 2','Legendary'], axis=1, inplace=True)
pokemon_clean_df.reset_index(drop=True, inplace=True)
pokemon_clean_df
```

### Multicollinearity check
- No correlations over 0.75 apart from itself.
- Attack, Special Attack and Special Defence are highly correlated with our dependant variable

```python
stat_predictors = pokemon_clean_df.iloc[:,1:8]
pd.plotting.scatter_matrix(stat_predictors, figsize  = [20, 20]);
plt.show()
```

```python
stat_predictors.corr()
```

### Normalise

```python
stat_predictors.hist(figsize=(15,10));
```

```python
columns = stat_predictors.columns
for c in columns:
    sm.graphics.qqplot(stat_predictors[c], dist=stats.norm, line='45', fit=True)
```

```python
# get JB value to see if data is normal distributed

# from scipy.stats import jarque_bera

# result = (jarque_bera(stat_predictors['Speed']))

# print(f"JB statistic: {result[0]}")
# print(f"p-value: {result[1]}")
```

```python
columns = stat_predictors.columns
data_log = pd.DataFrame([])
for c in columns:
    data_log[c] = np.log(stat_predictors[c])
data_log.hist(figsize=(15,10));
```

```python
for c in columns:
    sm.graphics.qqplot(data_log[c], dist=stats.norm, line='45', fit=True)
```

### One hot encode

```python
# dummy variables
type_dum = pd.get_dummies(pokemon_clean_df['Type 1'], prefix='type', drop_first=True)
gen_dum = pd.get_dummies(pokemon_clean_df['Generation'], prefix='gen', drop_first=True)

# add dummy variables
pokemon_log_df = pd.concat([data_log, type_dum, gen_dum], axis=1) # using log data
pokemon_df = pd.concat([stat_predictors, type_dum, gen_dum], axis=1) # using non-log data
pokemon_df2 = pd.concat([stat_predictors, gen_dum], axis=1) # remove type 1
```

### Modeling

```python
total = pokemon_df['Total']
predictors = pokemon_df2.drop(['Total'], axis=1)
```

#### Model 1

```python
predictors
```

```python
# using all predictors
predictors_int = sm.add_constant(predictors)
model = sm.OLS(total, predictors_int).fit()
model.summary()
```

```python
sm.graphics.qqplot(model.resid, dist=stats.norm, line='45', fit=True)
```
